```python
import pandas as pd

# fields
# company,pdf_source,year

df = pd.read_csv("../spreadsheet/climate_arc_matches.csv")
df.head(2)
```

```python
data = []
for index, row in df.iterrows():
    company_name = row["RTZ_name"]
    report_url = row[f"Source URL"]
    year = row["Latest emissions year"]
    data.append({"company_name": company_name, "report_url": report_url, "year": year})

# Convert the list of dictionaries to a DataFrame and save it as a CSV file
output_df = pd.DataFrame(data)
output_df = output_df.dropna()
output_df["year"] = output_df["year"].astype(int)
output_df.to_csv("data/sheetreports.csv", index=False)
```

```python
import json

# Load JSON data
with open('data/tech_reports_links.json', 'r') as file:
    json_data = json.load(file)

# Extract required fields
tech_data = []
for key, item in json_data.items():
    company_name = key
    report_url = item.get("pdf_link")
    year = item.get('year_published')
    tech_data.append({"company_name": company_name, "report_url": report_url, "year": year})

# Convert the list of dictionaries to a DataFrame and save it as a CSV file
tech_df = pd.DataFrame(tech_data)
tech_df = tech_df.dropna()
tech_df["year"] = tech_df["year"].astype(int)
tech_df.to_csv("data/tech_reports.csv", index=False)
```

```python
# merge
csvs = ["data/sheetreports.csv", "data/tech_reports.csv"]
# concatenate csvs into one dataframe
dfs = [pd.read_csv(csv) for csv in csvs]
merged_df = pd.concat(dfs, ignore_index=True)
merged_df.to_csv("data/reports_links.csv", index=False)
```
